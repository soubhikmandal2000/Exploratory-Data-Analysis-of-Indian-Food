```python
top_5=pd.crosstab(df["state"],["name"]).sort_values(by=["name","state"],ascending=False)
top_5.head(5).plot(kind="bar",color="pink",figsize=(8,8))
plt.title("TOP 5 STATES WITH MAX NUMBER OF DISHES")
plt.show()
```

```python
pd.crosstab(df["state"],df["diet"]).plot(kind="bar",figsize=(8,8))
plt.title("STATE WISE MOST PREFFERD TYPE OF DIET")
plt.show()
```

```python
pd.crosstab(df["state"],df["course"]).plot(kind="bar",figsize=(8,8))
plt.title("STATE WISE MOST PREFFERD COURSE")
plt.show()
```

```python
pd.crosstab(df["state"],df["flavor_profile"]).plot(kind="barh",figsize=(8,8))
plt.title("STATE WISE MOST PREFFERD FLAVOR")
plt.show()
```

```python
sns.countplot(df["diet"])
plt.title("WHAT DIET IS BEING MORE PREFFERED")
plt.show()
```

```python
sns.boxplot(x="diet",y="cook_time",data=df)
```

```python
df.groupby(df["diet"])["cook_time"].mean().plot(kind="bar",color="orange",figsize=(8,8))
plt.title("AVERAGE TIME TO COOK THE DIET")
plt.show()
```

```python
df.groupby(df["diet"])["prep_time"].mean().plot(kind="bar",color="blue",figsize=(8,8))
plt.title("AVERAGE TIME TO PREPARE DIET")
plt.show()
```

```python
pd.crosstab(df["diet"],df["flavor_profile"]).plot(kind="bar",figsize=(8,8))
plt.title("NUMBER OF FLAVOR-PROFILE IN DIET")
plt.show()
```

```python
pd.crosstab(df["diet"],df["course"]).plot(kind="bar",figsize=(8,8))
plt.title("NUMBER OF COURSES IN DIET")
plt.show()
```

```python
pd.crosstab(df["region"],df["diet"]).plot(kind="bar",figsize=(8,8))
plt.title("DIET BASED ON REGION")
plt.show()
```

```python
pd.crosstab(df["region"],df["course"]).plot(kind="bar",figsize=(8,8))
plt.title("COURSE BASED ON REGION")
plt.show()
```

```python
pd.crosstab(df["region"],df["flavor_profile"]).plot(kind="bar",figsize=(8,8))
plt.title("FLAVOR-PROFILE BASED ON REGION")
plt.show()
```
